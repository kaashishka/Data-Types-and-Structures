Ques 4: Describe how dictionaries store data?
- Ans: Dictionaries in Python store data using a structure called a hash table. This structure allows for efficient retrieval of values based on their associated keys. Here's a breakdown of how it works:
1.	Key-Value Pairs: Dictionaries store data in key-value pairs. Each key is unique and immutable (meaning it cannot be changed after creation). Values, on the other hand, can be of any data type and can be duplicated.
2.	Hashing: When a key-value pair is added to a dictionary, the key is passed through a hash function. This function generates a hash value, which is an integer that acts as an index into an array (the hash table).
3.	Storage in the Hash Table: The hash value determines the location (or "bucket") in the hash table where the value will be stored. Ideally, each key maps to a unique bucket. However, different keys can sometimes produce the same hash value, a situation known as a collision.
4.	Collision Handling: Various techniques are employed to handle collisions
Retrieval: To retrieve a value, Python again hashes the key and uses the hash value to locate the bucket in the hash table. If the bucket contains multiple values (due to a collision), additional checks (using the original key) are performed to retrieve the correct value.

Ques 8: What is a hash table, and how does it relate to dictionaries in Python?
- Ans: A hash table is a data structure that stores data in an associative manner, using key-value pairs. It employs a hash function to compute an index (or hash code) from each key. This index determines where the corresponding value is stored within the hash table's underlying array. This allows for efficient retrieval of values given their associated keys.

Ques 13: How do sets handle duplicate values in Python?
- Ans: Sets in Python are fundamentally designed to store only unique elements. When you attempt to add a duplicate value to a set, Python simply ignores the addition, and the set remains unchanged. It doesn't raise an error; the duplicate is silently discarded. This behavior is a core feature of sets and is what makes them useful for tasks like removing duplicates from a list or ensuring that all elements in a collection are unique.

Ques 17: Describe the time complexity of accessing elements in a dictionary
- Ans: Most common dictionary operations like retrieving by key, checking key existence, updates, and deletions have an average time complexity of O(1), making them extremely efficient. However, it's crucial to be aware of the theoretical worst-case O(n) scenario due to hash collisions, even though it's unlikely to occur in practice with Python's well-optimized implementation. Iterating and copying a dictionary always have a time complexity of O(n).
